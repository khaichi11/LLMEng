# Studi Kasus: Overfitting, Underfitting, dan Good Fit
## Analogi Kelompok Siswa dalam Ujian Matematika

Dalam studi kasus ini, kita menggunakan analogi tiga kelompok siswa:

- **Kelompok UF (Underfitting)** → Siswa malas belajar, tidak memahami konsep.
- **Kelompok OF (Overfitting)** → Siswa menghafal soal dan jawaban tanpa memahami konsep.
- **Kelompok GF (Good Fit)** → Siswa memahami konsep dan mampu menerapkannya.

Kita akan mensimulasikan kondisi berikut:

- **Kuis Harian** → Dianalogikan sebagai *Training Set*  
- **Ujian Akhir Semester** → Dianalogikan sebagai *Test Set*

Tujuan notebook ini adalah menunjukkan bagaimana:
- Overfitting terlihat bagus di training tetapi buruk di test.
- Underfitting buruk di keduanya.
- Good fit seimbang dan mampu generalisasi.

## 1. Import Library
Kita akan menggunakan dataset sintetis dan model sederhana dari scikit-learn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

## 2. Membuat Dataset Sintetis

Bayangkan dataset ini sebagai kumpulan soal matematika dengan fitur tertentu.
Kita buat pola non-linear agar bisa melihat efek overfitting dan underfitting.

```python
# Buat data sintetis
np.random.seed(42)

X = np.linspace(-3, 3, 200).reshape(-1, 1)
y = (X[:, 0]**2 + np.random.normal(0, 1, 200) > 4).astype(int)

# Split menjadi Training dan Test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## 3. Kelompok UF (Underfitting)

Kelompok ini seperti siswa yang tidak belajar.
Kita gunakan model yang terlalu sederhana (max_depth=1).

```python
model_uf = DecisionTreeClassifier(max_depth=1, random_state=42)
model_uf.fit(X_train, y_train)

train_acc_uf = accuracy_score(y_train, model_uf.predict(X_train))
test_acc_uf = accuracy_score(y_test, model_uf.predict(X_test))

print("Underfitting Model")
print("Training Accuracy:", round(train_acc_uf, 2))
print("Test Accuracy:", round(test_acc_uf, 2))
```

## 4. Kelompok OF (Overfitting)

Kelompok ini menghafal semua soal.
Kita gunakan model sangat kompleks (tanpa batas kedalaman).

```python
model_of = DecisionTreeClassifier(random_state=42)
model_of.fit(X_train, y_train)

train_acc_of = accuracy_score(y_train, model_of.predict(X_train))
test_acc_of = accuracy_score(y_test, model_of.predict(X_test))

print("Overfitting Model")
print("Training Accuracy:", round(train_acc_of, 2))
print("Test Accuracy:", round(test_acc_of, 2))
```

## 5. Kelompok GF (Good Fit)

Kelompok ini memahami konsep.
Kita gunakan model dengan kompleksitas sedang (max_depth=3).

```python
model_gf = DecisionTreeClassifier(max_depth=3, random_state=42)
model_gf.fit(X_train, y_train)

train_acc_gf = accuracy_score(y_train, model_gf.predict(X_train))
test_acc_gf = accuracy_score(y_test, model_gf.predict(X_test))

print("Good Fit Model")
print("Training Accuracy:", round(train_acc_gf, 2))
print("Test Accuracy:", round(test_acc_gf, 2))
```

## 6. Kesimpulan Hasil

Bandingkan hasil ketiga model:

- Jika Training dan Test sama-sama rendah → Underfitting
- Jika Training tinggi tapi Test rendah → Overfitting
- Jika Training dan Test sama-sama tinggi dan seimbang → Good Fit

Analogi dengan siswa:

- UF → Tidak belajar → Nilai kuis dan ujian sama-sama buruk
- OF → Menghafal → Nilai kuis tinggi, ujian akhir rendah
- GF → Paham konsep → Nilai kuis dan ujian sama-sama baik

Tujuan Machine Learning adalah menjadi seperti Kelompok GF.

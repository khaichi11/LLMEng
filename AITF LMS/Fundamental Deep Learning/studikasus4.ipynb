## 1. Variabel dan Fungsi

Penjelasan :
- Variabel: Simbol untuk menyimpan nilai yang bervariasi atau dapat berubah.
- Fungsi: Aturan yang mendefinisikan hubungan antara variabel input dan output, ditulis sebagai $y=f(x)$.
- Multivariable Function: Fungsi dengan lebih dari satu variabel input, contohnya $y=f(x_1, x_2, x_3)$.
- Composite Function (Fungsi Komposit): Fungsi yang bergantung pada fungsi lain.Contoh: Harga $h = f_1(k, d) = 10000 \times k - d$, di mana diskon $d = f_2(t) = 100 \times t$.

```python
# Definisi fungsi komposit
def f2(t):
    """Menghitung nilai diskon berdasarkan tanggal"""
    return 100 * t

def f1(k, d):
    """Menghitung harga akhir berdasarkan kuantitas dan diskon"""
    return (10000 * k) - d

# Implementasi
k = 5  # kuantitas
t = 10 # tanggal

d = f2(t)
h = f1(k, d)

print(f"Nilai diskon (d) = {d}")
print(f"Harga pembelian (h) = {h}")
```

## 2. Struktur dan Operasi pada Data (Skalar, Vektor, dan Matriks)

Penjelasan: 
- Skalar: Nilai yang hanya memiliki besaran (contoh: suhu 37.5).
- Vektor: Serangkaian skalar yang memiliki urutan, besaran, dan arah. Bisa berupa vektor baris atau vektor kolom (menggunakan operasi Transpose $T$).
- Matriks: Susunan skalar dengan baris ($m$) dan kolom ($n$), berukuran $m \times n$. 
- Dataset tabular dalam ML umumnya disimpan sebagai matriks.Kita dapat melakukan operasi aljabar linier seperti perkalian dot product, penjumlahan, pengurangan, dan perkalian skalar.

```python
import numpy as np

# Mendefinisikan Vektor dan Matriks
a = np.array([[1], [2]]) # Vektor kolom
B = np.array([[3, 5], [4, 6]]) # Matriks 2x2
C = np.array([[7, 9], [8, 10]]) # Matriks 2x2

print("Vektor a:\n", a)
print("\nTranspose Vektor a:\n", a.T)

# Operasi Matematika
print("\n1. Penjumlahan Matriks (B + C):\n", B + C)
print("\n2. Pengurangan Matriks (B - C):\n", B - C)
print("\n3. Perkalian Skalar (10 * B):\n", 10 * B)

# Dot Product (Perkalian Vektor baris dengan Matriks)
# Catatan: Syarat dot product, kolom matriks 1 = baris matriks 2
# Pada contoh teks, a transpos (baris) dikali B: [1, 2] * [[3,5], [4,6]]
dot_product = np.dot(a.T, B) 
print("\n4. Dot Product (a.T * B):\n", dot_product)
```

## 4. Turunan (Derivative) dan Kemiringan Kurva

Penjelasan:

**Turunan:** Laju perubahan nilai output terhadap perubahan input yang sangat kecil.  
Secara geometris, ini adalah kemiringan (gradien) garis singgung (*Tangent Line*) pada satu titik di kurva.

**Chain Rule (Aturan Rantai):** Digunakan untuk mencari turunan fungsi komposit:

$$
\frac{dy}{dx} = \frac{dy}{du} \times \frac{du}{dx}
$$

**Turunan Parsial:** Mencari turunan fungsi multivariable secara terpisah, di mana saat satu variabel diturunkan, variabel lain dianggap konstanta:

$$
\partial
$$

```python
import sympy as sp

# Mendefinisikan simbol
x, y = sp.symbols('x y')

# 1. Turunan fungsi tunggal: y = 0.5 * x^2
f_x = 0.5 * x**2
turunan_f = sp.diff(f_x, x)
print("Fungsi f(x):", f_x)
print("Turunan f'(x):", turunan_f)
print("Kemiringan di x = -2:", turunan_f.subs(x, -2)) # Substitusi x = -2

print("-" * 30)

# 2. Turunan Parsial (Multivariable): z = x^2 + y^3
z = x**2 + y**3
turunan_z_terhadap_x = sp.diff(z, x)
turunan_z_terhadap_y = sp.diff(z, y)
print("Fungsi z(x,y):", z)
print("Turunan parsial terhadap x (dz/dx):", turunan_z_terhadap_x)
print("Turunan parsial terhadap y (dz/dy):", turunan_z_terhadap_y)
```

## 5. Pencarian Titik Terdalam (Dasar Gradient Descent)

Penjelasan:

- Gradien digunakan sebagai penunjuk arah untuk mencari nilai minimum suatu fungsi (titik terdalam kurva).

- Jika gradien negatif (kurva turun ke kanan), gerakkan variabel input $x$ ke arah positif (kanan).

- Jika gradien positif (kurva naik ke kanan), gerakkan variabel input $x$ ke arah negatif (kiri).

- Berhenti jika gradien mencapai 0 (titik minimum).

- Ini adalah konsep fundamental dari Gradient Descent.



```python
# Simulasi sederhana Gradient Descent untuk fungsi y = x^2
# Turunannya adalah dy/dx = 2x

x_current = 4.0 # Titik awal tebakan acak
learning_rate = 0.1 # Seberapa besar langkah yang diambil
epochs = 15 # Jumlah iterasi pencarian

print(f"Nilai awal x = {x_current}")

for i in range(epochs):
    # Hitung gradien saat ini (2x)
    gradient = 2 * x_current 
    
    # Update nilai x berlawanan dengan arah gradien
    x_current = x_current - (learning_rate * gradient)
    
    print(f"Iterasi {i+1}: Gradien = {gradient:.4f}, Update x = {x_current:.4f}")

print(f"\nNilai minimum tercapai di sekitar x = {x_current:.4f}")
```

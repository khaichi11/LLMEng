# Optimasi dan Regularisasi

Pada modul ini, kita akan melakukan eksperimen pembuatan program Deep Learning untuk mengidentifikasi gambar dari dataset Fashion MNIST. Kita juga akan mencoba beberapa eksperimen seperti penambahan jumlah *hidden layer*, penggantian *optimizer* (SGD vs Adam), serta penggunaan teknik regularisasi (Dropout).

[Image of Fashion MNIST dataset samples]

### 1. Import Library dan Persiapan Dataset
Langkah pertama adalah mengimpor *library* TensorFlow dan Keras yang dibutuhkan. Pastikan Anda menggunakan GPU agar waktu eksekusi pelatihan model dapat dipersingkat.

Selanjutnya, kita akan mengunduh dataset `fashion_mnist`, membaginya menjadi data latih (*Training Set*) dan data uji (*Test Set*), lalu melakukan normalisasi nilai piksel pada rentang 0 вҖ“ 1.

```python
from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dropout,Dense
from tensorflow import keras

fashion = datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion.load_data() 

X_train = X_train/255.0
X_test  = X_test/255.0
```

### 2. Eksperimen 1: ANN dengan 1 Hidden Layer dan SGD Optimizer



Pada eksperimen pertama ini, kita menyiapkan model *Artificial Neural Network* (ANN) sederhana dengan 1 *hidden layer* yang memiliki 128 *node*. 
* **Optimizer:** Stochastic Gradient Descent (SGD)
* **Loss Function:** sparse_categorical_crossentropy
* **Epochs:** 100
* **Validation Split:** 20% dari Training Set

```python
model1 = keras.models.Sequential()
model1.add(Flatten(input_shape=[28, 28]))
model1.add(Dense(128, activation='relu'))
model1.add(Dense(10, activation='softmax'))

model1.compile(loss='sparse_categorical_crossentropy', 
optimizer='SGD',
metrics=['accuracy'])

model1.summary()

model1.fit(X_train, y_train, epochs=100, 
validation_split=0.2, 
batch_size=128)

model1.evaluate(X_test, y_test)
```

### 4. Eksperimen 3: DNN dengan 2 Hidden Layer dengan Adam Optimizer

Arsitektur model ini persis sama dengan arsitektur `model2` (2 *hidden layer* dengan masing-masing 128 *node*). Perbedaannya terletak pada penggunaan **Adam optimizer** sebagai pengganti SGD untuk melihat efisiensi dan peningkatan performa pelatihannya.

```python
model3 = keras.models.Sequential()
model3.add(Flatten(input_shape=[28, 28]))
model3.add(Dense(128, activation='relu'))
model3.add(Dense(128, activation='relu'))
model3.add(Dense(10, activation='softmax'))

model3.compile(loss='sparse_categorical_crossentropy', 
optimizer='Adam',
metrics=['accuracy'])

model3.summary()

model3.fit(X_train, y_train, epochs=100, 
validation_split=0.2, 
batch_size=128)

model3.evaluate(X_test, y_test)
```

### 5. Eksperimen 4: DNN dengan 2 Hidden Layer dengan Adam Optimizer dan Dropout



Untuk mencegah masalah *overfitting*, kita mencoba teknik regularisasi dengan menambahkan **Dropout layer** (sebesar 0.3 atau 30%) masing-masing setelah *input layer* dan setiap *hidden layer*. Dropout mematikan sebagian *node* secara acak selama proses pelatihan. Tidak ada Dropout yang ditambahkan pada *output layer*.

```python
model4 = keras.models.Sequential()
model4.add(Flatten(input_shape=[28, 28]))
model4.add(Dropout(0.3))
model4.add(Dense(128, activation='relu'))
model4.add(Dropout(0.3))
model4.add(Dense(128, activation='relu'))
model4.add(Dropout(0.3))
model4.add(Dense(10, activation='softmax'))

model4.compile(loss='sparse_categorical_crossentropy', 
optimizer='Adam',
metrics=['accuracy'])

model4.summary()

model4.fit(X_train, y_train, epochs=100, 
validation_split=0.2, 
batch_size=128)

model4.evaluate(X_test, y_test)
```

### 6. Eksperimen Penambahan Jumlah Layer (6 dan 8 Hidden Layer)

Dengan menggunakan Keras `Sequential` API, penambahan jumlah *hidden layer* dapat dilakukan dengan mudah hanya dengan menambahkan `model.add(Dense(...))` berulang kali sesuai dengan kedalaman arsitektur yang diinginkan. Berikut adalah perbandingan kode model dengan 6 dan 8 *hidden layer*.

```python
# DNN dengan 6 Hidden layer
model6 = keras.models.Sequential()
model6.add(Flatten(input_shape=[28, 28]))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(128, activation='relu'))
model6.add(Dense(10, activation='softmax'))

# DNN dengan 8 Hidden layer
model8 = keras.models.Sequential()
model8.add(Flatten(input_shape=[28, 28]))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(128, activation='relu'))
model8.add(Dense(10, activation='softmax'))

# Kompilasi dan pelatihan pada model8 sebagai contoh:
model8.compile(loss='sparse_categorical_crossentropy', 
optimizer='Adam',
metrics=['accuracy'])

model8.fit(X_train, y_train, epochs=100, 
validation_split=0.2,
batch_size=128)
model8.evaluate(X_test, y_test)
```

### Penjelasan: Apa itu Tensor?



Jika kita ingin mesin PyTorch memproses sesuatu (baik itu gambar, suara, atau teks), kita harus menerjemahkannya dulu ke dalam bahasa yang dimengerti oleh PyTorch. Bahasa itu disebut **Tensor**.

Secara sederhana, Tensor hanyalah sekumpulan angka (matriks). 
- Gambar hitam putih ukuran 2x2 pixel akan diubah menjadi Tensor berisi 4 angka.
- Gambar berwarna akan diubah menjadi tumpukan Tensor yang lebih banyak.

Jadi, ketika nanti kita melihat perintah `transforms.ToTensor()` pada saat memuat gambar wajah DAiSEE, tugas utamanya hanyalah mengubah gambar JPG/PNG menjadi deretan angka seperti yang kita buat di atas, agar bisa dihitung oleh rumus matematika.

```python
# 1. Ini adalah list (daftar angka) biasa di dalam bahasa pemrograman Python
angka_python = [1, 2, 3, 4]
print("Bentuk angka Python biasa :", angka_python)

# 2. Kita mengubah angka biasa tersebut menjadi format yang dipahami PyTorch
# Format ini dinamakan "Tensor"
angka_tensor = torch.tensor(angka_python)
print("Bentuk angka di PyTorch   :", angka_tensor)

# 3. Kenapa harus diubah ke Tensor? Karena Tensor bisa dihitung menggunakan GPU (VGA)
# Mari kita buat tensor berisi angka desimal acak sebanyak 5 buah
tensor_acak = torch.rand(5)
print("Tensor acak buatan PyTorch:", tensor_acak)
```

### 2. Berkenalan dengan `torch.nn`

Di PyTorch, semua mesin pembuat keputusan (layer) disimpan di dalam sebuah kotak peralatan khusus bernama `torch.nn` (*nn* singkatan dari *Neural Network*). 



Mari kita ambil mesin yang paling dasar, yaitu **`nn.Linear`**. Tugas mesin ini sederhana: menerima sejumlah angka, mengalikannya dengan "bobot" rahasia yang ia miliki, dan mengeluarkan angka baru sebagai kesimpulan.

```python
# 1. Memanggil kotak peralatan khusus mesin JST
import torch.nn as nn

# 2. Kita buat bahan baku (Tensor) yang berisi 3 angka
# Bayangkan ini adalah 3 pixel dari sebuah gambar
bahan_baku = torch.tensor([1.0, 2.0, 3.0])
print("1. Bahan baku yang masuk :", bahan_baku)

# 3. Mesin Linear dari PyTorch
# in_features=3 (karena bahan baku kita terdiri dari 3 angka)
# out_features=2 (kita ingin mesin memerasnya menjadi 2 angka kesimpulan saja)
mesin_linear = nn.Linear(in_features=3, out_features=2)

# 4. Masukkan bahan baku ke dalam mesin
hasil_proses = mesin_linear(bahan_baku)
print("2. Hasil setelah diproses:", hasil_proses)
```

### Penjelasan: Apa yang baru saja terjadi?

Perhatikan hasil cetakannya. Kita memasukkan 3 angka `[1.0, 2.0, 3.0]`, tapi keluarnya menjadi 2 angka acak yang berbeda (misalnya `[0.45, -1.2]`). 

Dari mana angka acak itu berasal? Saat kita mendefinisikan `nn.Linear`, mesin tersebut secara otomatis membuat nilai "bobot" (*weight*) secara acak di dalam dirinya. Ia mengalikan bahan baku kita dengan bobot acak tersebut. 

Nanti, saat proses *training* (pelatihan) berlangsung, nilai bobot acak inilah yang akan terus diperbaiki dan dikalibrasi oleh komputer agar tebakannya menjadi akurat!

```python
# 1. ReLU
mesin_penyaring = nn.ReLU()

# 2. Misalkan dari mesin Linear sebelumnya, kita mendapat angka negatif
tensor_dengan_negatif = torch.tensor([-1.5, 2.0, -0.5, 3.0])
print("Sebelum disaring:", tensor_dengan_negatif)

# 3. Kita lewatkan ke mesin penyaring
hasil_saringan = mesin_penyaring(tensor_dengan_negatif)
print("Setelah disaring :", hasil_saringan)
```

### Penjelasan: Mengapa kita butuh penyaring ReLU?

Seperti yang terlihat pada hasil *print*, cara kerja **ReLU** (Rectified Linear Unit) sangat amat sederhana: 
- Jika angkanya minus (negatif), langsung ubah jadi `0`.
- Jika angkanya plus (positif), biarkan saja lewat.



Pertanyaannya: **Buat apa mengubah angka negatif jadi nol?**
Jaringan Syaraf Tiruan butuh fungsi penyaring seperti ini agar pemikirannya tidak kaku (dalam matematika disebut non-linear). Tanpa ReLU, seberapa banyak pun lapisan `nn.Linear` yang kita tumpuk, JST kita tidak akan pernah bisa mengenali pola yang rumit seperti bentuk mata atau lengkungan senyum pada wajah. ReLU membantu JST membuang informasi yang dianggap "tidak berguna" (angka negatif) dan fokus pada sinyal yang kuat (angka positif).

### 3. Menggabungkan Layer menjadi Jaringan Utuh (`nn.Module`)

Setiap kali kita mau membuat model *deep learning* di PyTorch (baik itu model sederhana maupun model raksasa seperti Swin Transformer), kita **wajib** membuat sebuah *Class* yang mewarisi sifat dari `nn.Module`.



Di dalam *Class* ini, aturannya sangat baku dan selalu ada dua fungsi utama yang harus kita tulis:
1. **`__init__`**: Di sinilah kita mendefinisikan dan mendaftar layer apa saja yang mau kita pakai.
2. **`forward`**: Di sinilah kita mengatur urutan jalannya data. Layer mana yang memproses data pertama kali, kedua, dan seterusnya.

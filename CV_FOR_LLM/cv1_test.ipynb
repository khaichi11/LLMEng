# Realtime Object Detection Menggunakan YOLO dan RT-DETR

Notebook ini menjelaskan implementasi deteksi objek secara realtime menggunakan webcam dengan dua model berbeda, yaitu YOLO dan RT-DETR.  
Pengguna dapat mengganti model selama program berjalan untuk membandingkan performa dan kecepatan inferensi.

Kontrol keyboard:
- Tekan `1` untuk menggunakan YOLO
- Tekan `2` untuk menggunakan RT-DETR
- Tekan `q` untuk keluar dari program

## Import Library

Library yang digunakan:
- cv2 → untuk akses webcam dan menampilkan video
- ultralytics.YOLO → untuk memuat dan menjalankan model deteksi
- time → untuk menghitung FPS

```python
import cv2
from ultralytics import YOLO
import time
```

## Load Model

Dua model dimuat di awal program:

- yolo26n.pt → model ringan dan cepat
- rtdetr-l.pt → model berbasis transformer dengan ukuran lebih besar

Model default yang digunakan saat awal program adalah YOLO.

```python
yolo_model = YOLO("yolo26n.pt")
rtdetr_model = YOLO("rtdetr-l.pt")

current_model = yolo_model
model_name = "YOLO"
```

## Inisialisasi Webcam

Program menggunakan webcam default (index 0).
Jika menggunakan kamera eksternal, index dapat diganti sesuai kebutuhan.

```python
cap = cv2.VideoCapture(0)
```

## Proses Deteksi Realtime

Pada bagian ini program akan:
1. Mengambil frame dari webcam
2. Menjalankan model yang sedang aktif
3. Menggambar bounding box hasil deteksi
4. Menghitung FPS
5. Menampilkan hasil ke layar
6. Mengecek input keyboard untuk mengganti model atau keluar

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    start = time.time()

    results = current_model(frame, verbose=False)
    annotated_frame = results[0].plot()

    end = time.time()
    fps = 1 / (end - start)

    cv2.putText(annotated_frame, f"Model: {model_name}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

    cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 70),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

    cv2.imshow("Realtime Detection", annotated_frame)

    key = cv2.waitKey(1) & 0xFF

    if key == ord('1'):
        current_model = yolo_model
        model_name = "YOLO"

    elif key == ord('2'):
        current_model = rtdetr_model
        model_name = "RT-DETR"

    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## Penjelasan

Program ini dapat digunakan untuk:
- Menguji performa model deteksi objek secara realtime
- Membandingkan FPS antara dua arsitektur berbeda
- Melakukan demonstrasi computer vision berbasis webcam

### Belajar API LLM
Disini apabila ingin memanggil API LLM seperti GPT kita lebih baik menggunakan file .env yang tugasnya untuk menyimpan environment variable secara aman agar ketika repo di upload API key-nya tidak ikut tersebar. Berikut adalah kode program untuk mengambil API key di python menggunakan .env tersebut. 

```python
import os
from dotenv import load_dotenv
load_dotenv() #untuk membaca file .env
api_key = os.getenv("OPENAI_API_KEY") #mengambil nilai dari variabel lingkungan OPENAI_API_KEY
```

Selanjutnya hubungkan dengan `pip install openai` lalu lanjutkan dengan melakukan coding pada python

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)
```

Lakukan prompting ke model dengan format memberikan nama model yang akan digunakan dan memberikan pesan apa yang akan dikirimkan



Untuk melakukan prompting pada model bahasa (LLM), sebaiknya menggunakan struktur pesan berbentuk list of dictionaries seperti berikut:

`[
  {"role": "system", "content": "..."},
  {"role": "user", "content": "..."}
]`

Struktur ini digunakan untuk mensimulasikan percakapan antara pengguna dan model.

Setiap elemen dalam list memiliki dua komponen utama:

1. role → Menunjukkan peran dalam percakapan.
2. content → Berisi teks atau pesan yang dikirim.

Penjelasan masing-masing role:

- system  
  Digunakan untuk memberikan instruksi global kepada model, seperti mengatur gaya bahasa, kepribadian, batasan jawaban, atau konteks khusus yang harus diikuti model selama percakapan.

- user  
  Berisi pertanyaan atau perintah dari pengguna yang ingin dijawab oleh model.

- assistant (opsional)  
  Digunakan untuk menyimpan jawaban model sebelumnya dalam percakapan multi-turn agar konteks tetap terjaga.

Dengan menggunakan struktur ini, model dapat menghasilkan respons yang lebih terkontrol, konsisten, dan sesuai dengan konteks yang diinginkan.


```python
reponse = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Jelaskan apa itu AI secara singkat."}
    ]
)

print(reponse.choices[0].message.content)
```

Berikut adalah kode dari chatbot sederhana:

```python
pertanyaan = input("Tanya sesuatu:")

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": pertanyaan}
    ]
)

print(response.choices[0].message.content)
```

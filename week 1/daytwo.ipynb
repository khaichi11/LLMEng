Pada materi sebelumnya, kita diajarkan untuk tidak pernah membiarkan kunci API kita tersebar.

- Fungsi `load_dotenv()` bertugas membaca file bernama .env yang ada di komputermu.

- Fungsi `os.getenv("OPENAI_API_KEY")` digunakan untuk mengambil nilai kunci rahasia tersebut dari lingkungan sistem (environment variable) secara aman.

```python
import os
from dotenv import load_dotenv

# Membaca file .env
load_dotenv(override=True) 
api_key = os.getenv('OPENAI_API_KEY')

if api_key:
    print("API key berhasil dimuat!")
else:
    print("API key tidak ditemukan, cek file .env kamu.")
```

Tidak seperti sebelumnya yang di mana kita langsung memanggil API menggunakan library openai, sekarang kita akan menggunakan konsep Endpoint untuk mengetahui bagaimana API dipanggil seperti pada library:

- Endpoint adalah alamat URL (Uniform Resource Locator) web tempat kita mengirimkan data.

- Kode di bawah ini menggunakan fungsi `requests.post()` untuk mengirimkan instruksi secara manual langsung ke server OpenAI di alamat `https://api.openai.com/v1/chat/completions`.

- Data yang dikirim dibungkus dalam variabel payload berformat JSON.

```python
import requests

# Ini adalah cara manual (mentah) memanggil AI menggunakan Endpoint HTTP
headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model": "gpt-4.1-mini", 
    "messages": [
        {"role": "user", "content": "Tell me a fun fact"}
    ]
}

response = requests.post(
    "https://api.openai.com/v1/chat/completions",
    headers=headers,
    json=payload
)

# Mengekstrak teks balasan dari format JSON yang rumit
print(response.json()["choices"][0]["message"]["content"])
```

Menulis format requests.post() dan mengekstrak JSON secara manual seperti di Sel 2 itu sangat merepotkan, Karena itu package openai (yang kita install via pip sebelumnya) masuk.

- Package ini hanyalah sebuah Python Client Library atau bungkusan (wrapper) untuk melakukan panggilan HTTP Endpoint secara otomatis.

- Penting diingat: Library ini open-source, sangat ringan (lightweight), dan sama sekali tidak mengandung kode pemrograman model AI di dalamnya!

```python
from openai import OpenAI

# Jauh lebih bersih dan rapi dibandingkan request HTTP manual!
client = OpenAI(api_key=api_key)

response = client.chat.completions.create(
    model="gpt-4.1-mini", 
    messages=[{"role": "user", "content": "Tell me a fun fact"}]
)

print(response.choices[0].message.content)
```

- Pada materi sebelumnya, kita membuat variabel messages berbentuk daftar dictionary (list of dictionaries).

- Kunci role: system digunakan untuk memberikan instruksi kepribadian (personality) atau batasan kepada AI, sedangkan role: user adalah pertanyaan dari kita.

- Di Day 2, ini disebut sebagai Chat Completions API, yang artinya kita memberikan percakapan ini ke sistem, lalu meminta sistem untuk memprediksi apa yang harus muncul selanjutnya (predict what should come next).

```python
# Mendefinisikan peran (role) dengan rapi
pesan_percakapan = [
    {"role": "system", "content": "You are a helpful and snarky assistant."},
    {"role": "user", "content": "Jelaskan apa itu Endpoint dengan gaya bahasa lucu."}
]

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=pesan_percakapan
)

print(response.choices[0].message.content)
```
